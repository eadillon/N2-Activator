```python
# load data
import os
from rewrap_adjlen2 import rewrap
atoms = 40 # number of atoms in input
size = atoms*4
training_data = rewrap('/Users/enricadillon/Python/data/terminal_zax_csv',length=size)
test_data = rewrap('/Users/enricadillon/Python/data/terminal_zax_testdata',length=size)
#validation_data = vector('/Users/enricadillon/Python/data/search4_validationdata')
training_data[0][0].shape
```

```python
#convert atomic number to periodic coordinates w electronegativity
os.chdir('/Users/enricadillon/Python/N2predicter') 
from periodic_coordinates_EN import AN_to_PC
training_data_PC = AN_to_PC(training_data)
test_data_PC = AN_to_PC(test_data)
test_data_PC[0][0].shape
```

```python
#import network
os.chdir('/Users/enricadillon/Python/N2predicter') 
from N2_neural_network import Network, MeanSquaredErrorCostWithConstraint
net = Network([int(size*1.5), 60, 60, 6],cost=MeanSquaredErrorCostWithConstraint) #structure of network: input,layer1,...,layern,output
net.default_weight_initializer()
```

```python
#run network
results = net.SGD(training_data_PC, 500, 650, 0.1, lmbda=1, lr_decay=0.999,
                  evaluation_data=test_data_PC,
                  monitor_evaluation_cost=False,
                  monitor_evaluation_accuracy=True,  
                  monitor_training_cost=False,
                  monitor_training_accuracy=True) #data, epochs, batch size, eta, lmbda, lr_decay
min(results[1])
```

```python
os.getcwd()
```

```python
import pandas as pd

# Save results to CSV
evaluation_accuracy = results[1]
training_accuracy = results[3]
epochs = list(range(1, len(evaluation_accuracy) + 1))

results_df = pd.DataFrame({
    'Epoch': epochs,
    'Evaluation Accuracy': evaluation_accuracy,
    'Training Accuracy': training_accuracy
})

results_df.to_csv('network_results.csv', index=False)

# Print the minimum evaluation accuracy
min(results[1])
```
